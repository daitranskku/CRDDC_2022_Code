```python
import os
import json
import matplotlib.pyplot as plt
from tqdm import tqdm
import cv2
```

```python
def get_json_data(label_path):
    with open(label_path, encoding='utf-8-sig') as f:
        json_data = json.load(f)
    return json_data

# temp is [x1,y1,x2,y2]
# bbox_list is [x1,y1,x2,y2]
def check_valid(bbox_list,w,h):
    temp = []
    
    # x1
    if bbox_list[0]<0:
        temp.append(0)
    else:
        temp.append(bbox_list[0])
        
    # y1
    if bbox_list[1]<0:
        temp.append(0)
    else:
        temp.append(bbox_list[1])
    
    # x2
    if bbox_list[2]>w:
        temp.append(w)
    else:
        temp.append(bbox_list[2])
    
    # y2
    if bbox_list[3]>h:
        temp.append(h)
    else:
        temp.append(bbox_list[3])
        
    return temp

def yolo_style(box,w,h):
    # center_x,center_y,w,h
    coor_x = (box[0] + box[2])/2.0
    coor_y = (box[1] + box[3])/2.0
    coor_w = box[2] - box[0]
    coor_h = box[3] - box[1]

    return (coor_x/w,coor_y/h,coor_w/w,coor_h/h)

    
    
```

```python
img_list=json_data['images']
ann_list=json_data['annotations']
cate_list=json_data['categories']

# train_ann_save = "/root/DataSet/darwin/si/crack_maskset/labels/"
# IMAGE_DIR= "/root/DataSet/darwin/si/crack_maskset/images/"

train_ann_save = "/SSD/IEEE/data/yolo_dataset/labels/train/"

if not os.path.exists(train_ann_save): 
    os.makedirs(train_ann_save)
```

### make yolo set

```python
for img_dic in img_list:
    img_id = img_dic['id']
    
    w = img_dic['width']
    h = img_dic['height']
    
    class_bbox = []
    check_bit = 0
    for ann_dic in ann_list:
        if ann_dic['image_id'] == img_id:
            
            check_bit += 1
            
            temp=[]
            
            cate_id=int(ann_dic['category_id'])
            cate_name = cate_id-1
            temp.append(cate_name)
            
            bbox=ann_dic['bbox']
            
            # x,y,w,h to x1,y1,x2,y2
            x1 = bbox[0]
            y1 = bbox[1]
            x2 = bbox[0]+bbox[2]
            y2 = bbox[1]+bbox[3]     
            
            # output is x1,y1,x2,y2
            valid_bbox=check_valid([x1,y1,x2,y2],w,h)
            
            # yolo style
            yolo_bbox = yolo_style(valid_bbox,w,h)
            
            for coor in yolo_bbox:
                temp.append(coor)
        
            class_bbox.append(temp)
            
        elif check_bit != 0:
            break
    
    file_name = img_dic['file_name'].split('.')[0]
#     save_name = file_name.split('.')[0]
    
#     # for V7
#     if not os.path.exists(train_ann_save+save_name.split('/')[0]): 
#         os.makedirs(train_ann_save+save_name.split('/')[0])
    
    with open(train_ann_save+file_name+'.txt','w',encoding='UTF-8') as f:
        for row in class_bbox:
            for col in row:
                f.write(str(col)+' ')
            f.write('\n')
                
    
```
