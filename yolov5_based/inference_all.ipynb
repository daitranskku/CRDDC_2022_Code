```python
def get_image_file_names(IMAGE_DIR):
    
    file_format = ['jpg']
    
    file_names = os.listdir(IMAGE_DIR)
    image_file_names = []
    for file_name in tqdm(file_names):
        if file_name.split('.')[-1] in file_format:
            image_file_names.append(file_name)
   
    return natsort.natsorted(image_file_names)
```

```python
def get_json_data(label_path):
    with open(label_path, encoding='utf-8-sig') as f:
        json_data = json.load(f)
    return json_data
```

```python
def get_iou(pred_box, gt_box):
    ixmin = max(pred_box[0], gt_box[0])
    ixmax = min(pred_box[2], gt_box[2])
    iymin = max(pred_box[1], gt_box[1])
    iymax = min(pred_box[3], gt_box[3])
    iw = np.maximum(ixmax-ixmin+1., 0.)
    ih = np.maximum(iymax-iymin+1., 0.)
    inters = iw*ih
    uni = ((pred_box[2]-pred_box[0]+1.) * (pred_box[3]-pred_box[1]+1.) +
           (gt_box[2] - gt_box[0] + 1.) * (gt_box[3] - gt_box[1] + 1.) -
           inters)
    iou = inters / uni
    return iou
```

```python
def visualize_pred_bboxes_size(pred_boxes, pred_labels, image):
    image_shape = image.shape
    for pred_label_num, pred_box in enumerate(pred_boxes):
        # print(pred_box)
        start_point = (int(pred_box[0]*image_shape[0]), int(pred_box[1]*image_shape[1]))
        end_point =   (int(pred_box[2]*image_shape[0]), int(pred_box[3]*image_shape[1]))
        class_id = pred_labels[pred_label_num]
        image = cv2.rectangle(image, start_point, end_point,  color_dict[class_id], 10)
        cv2.putText(image, class_dict_raw[class_id], start_point, 
                cv2.FONT_HERSHEY_SIMPLEX, 2, color_dict[class_id], 3, cv2.LINE_AA)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def visualize_pred_bboxes(pred_boxes, pred_labels, image):
    for pred_label_num, pred_box in enumerate(pred_boxes):
        # print(pred_box)
        start_point = (int(pred_box[0]), int(pred_box[1]))
        end_point =   (int(pred_box[2]), int(pred_box[3]))
        class_id = pred_labels[pred_label_num]
        image = cv2.rectangle(image, start_point, end_point,  color_dict[class_id], 10)
        cv2.putText(image, class_dict_raw[class_id], start_point, 
                cv2.FONT_HERSHEY_SIMPLEX, 2, color_dict[class_id], 3, cv2.LINE_AA)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def one_stage_convert2ensemble(model, image, IMAGE_SIZE, threshold = 0.2):
    results = model(image.copy(), size = IMAGE_SIZE)
    image_size = image.shape
    extracted_results = results.xyxy[0]
    pred_boxes = []
    pred_labels = []
    pred_scores = []
    for pred_num in range(len(extracted_results)):
        pred_score = extracted_results[pred_num][-2].item()
        if pred_score >= threshold:
            class_id = extracted_results[pred_num][-1].item()
            temp_box = [extracted_results[pred_num][0].item()/image_size[1],
                        extracted_results[pred_num][1].item()/image_size[0],
                        extracted_results[pred_num][2].item()/image_size[1],
                        extracted_results[pred_num][3].item()/image_size[0],
                       ]
            pred_score = extracted_results[pred_num][-2].item()
            pred_boxes.append(temp_box)
            pred_labels.append(int(class_id+1))
            pred_scores.append(pred_score)
    return pred_boxes, pred_labels, pred_scores
```

```python
def two_stage_convert2ensemble(model, image, threshold = 0.4):
    results = inference_detector(model, image)
    image_size = image.shape
    pred_boxes = []
    pred_scores = []
    pred_labels = []
    for predicted_class in range(len(results)):
        for preds in results[predicted_class]:
            if preds[-1] >= threshold:
                temp_box = preds[:4]
                pred_boxes.append([temp_box[0]/image_size[1],
                                   temp_box[1]/image_size[0],
                                   temp_box[2]/image_size[1],
                                   temp_box[3]/image_size[0]]
                                   )
                pred_score = preds[-1].astype(float)
                pred_scores.append(pred_score)
                pred_labels.append(int(predicted_class+1))
    return pred_boxes, pred_labels, pred_scores
```

```python
def ensemble_models(models, image, visualize = True):
    image_size = image.shape
    boxes_list_merges = []
    labels_list_merges = []
    scores_list_merges = []
    for i, model in enumerate(models):
        if model_types[i] == 1:
            pred_boxes, pred_labels, pred_scores = one_stage_convert2ensemble(model, image, IMAGE_SIZES[i],  ONE_STAGE_THESHOLDS[i])
            # print(ONE_STAGE_THESHOLDS[i])
        elif model_types[i] == 2:
            pred_boxes, pred_labels, pred_scores = two_stage_convert2ensemble(model, image, two_stage_model_threshold)
            # print(pred_labels)
        # Visualize each models
#         print("Model: {}".format(i))
#         visualize_pred_bboxes_size(pred_boxes, pred_labels, image.copy())
        # ensemble
        boxes_list_merges.append(pred_boxes)
        labels_list_merges.append(pred_labels)
        scores_list_merges.append(pred_scores)
    normalize_ensemble_boxes, ensemble_scores, normalize_ensemble_labels = weighted_boxes_fusion(boxes_list_merges, scores_list_merges, 
                                                      labels_list_merges, weights=weights, iou_thr=iou_thr, skip_box_thr=skip_box_thr)
    ensemble_boxes = []
    ensemble_labels = []
    for i, predicted_box in enumerate(normalize_ensemble_boxes):
        unnormalize_box = [predicted_box[0]*image_size[1],
                         predicted_box[1]*image_size[0],
                         predicted_box[2]*image_size[1],
                         predicted_box[3]*image_size[0]]
        ensemble_boxes.append(unnormalize_box)
        ensemble_labels.append(int(normalize_ensemble_labels[i]))
        
        if visualize == True:
            start_point = (int(unnormalize_box[0]), int(unnormalize_box[1]))
            end_point =   (int(unnormalize_box[2]), int(unnormalize_box[3]))
            image = cv2.rectangle(image, start_point, end_point,  color_dict[int(normalize_ensemble_labels[i])], 10)
            cv2.putText(image, class_dict_raw[int(normalize_ensemble_labels[i])], start_point, 
                        cv2.FONT_HERSHEY_SIMPLEX, 2, color_dict[int(normalize_ensemble_labels[i])], 3, cv2.LINE_AA)
    if visualize == True:
        plt.imshow(image)
        plt.show()
        
    return ensemble_boxes, ensemble_labels
```

## Submission

```python
def inference2submission(country = 'Norway'):
    TEST_SET_DIR = '/media/daitran/Data_SSD1/IEEE/data/{}/test/images'.format(country)
    test_image_file_names  = get_image_file_names(TEST_SET_DIR)
    submission = pd.DataFrame(columns=['ImageId', 'PredictionString'])
    for test_num, test_image_file_name in tqdm(enumerate(test_image_file_names)):
        test_image_path = os.path.join(TEST_SET_DIR, test_image_file_name)
        image = cv2.imread(test_image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image_size = image.shape

        ensemble_boxes, ensemble_labels = ensemble_models(models, image, visualize = False)
        # visualize
        # visualize_pred_bboxes(ensemble_boxes, ensemble_labels, image.copy())
        # Submission
        pred_str = ''
        for pred_num, pred_box in enumerate(ensemble_boxes):
            class_id = ensemble_labels[pred_num]
            temp_box = pred_box
            pred = '{} {} {} {} {} '.format(int(class_id),
                                            int(temp_box[0]),
                                            int(temp_box[1]),
                                            int(temp_box[2]),
                                            int(temp_box[3]))
            pred_str = pred_str + pred
        final_pred_str = pred_str[:-1] 
        submission.loc[test_num,'ImageId'] = test_image_file_name
        submission.loc[test_num,'PredictionString'] = final_pred_str
    # CONVERT TO TXT
    data = submission
    save_lines = []
    for row_num in range(len(data)):
        row = data.iloc[row_num]
        if pd.isnull(row['PredictionString']):
            temp_txt = '{},'.format(row['ImageId'])
        else:
            temp_txt = '{},{}'.format(row['ImageId'], row['PredictionString'])
        save_lines.append(temp_txt)
    with open('./results/{}_yolov5x.txt'.format(country), 'w') as f:
        for line in save_lines:
            f.write(line)
            f.write('\n')
    return submission
```
